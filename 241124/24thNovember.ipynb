Univariate:

"Uni" means one: You study just one thing at a time.
Example: Counting how many students are in each age group.
Focus: Distribution (How the values of one variable are spread out).

Bivariate:

"Bi" means two: You study two things to see if they are connected.
Example: Do students who study more hours get higher scores?
Focus: Relationship or comparison between two variables.

Multivariate:

"Multi" means many: You study three or more things together.
Example: Does studying more hours help both boys and girls improve their scores, and does age make a difference?
Focus: Complex relationships involving many variables.

```python
# Day one: age and speed of 13 cars
x1 = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y1 = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
plt.scatter(x1, y1, color='blue', label="Day One Data")

# Day two: age and speed of 15 cars
x2 = np.array([52, 52, 58, 51, 65, 58, 62, 59, 57, 53, 51, 54, 57, 54, 12])
y2 = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])
plt.scatter(x2, y2, color='green', label="Day Two Data")

# Add labels and title
plt.xlabel("Age of Cars")
plt.ylabel("Speed")
plt.title("Scatter Plot: Age vs Speed of Cars")
plt.legend()
plt.show()
```

```python
# Univariate Analysis - Histogram 
scores = [2, 5, 8, 12, 7, 6] 
score_ranges = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60'] 
plt.figure(figsize=(10, 5)) 
plt.bar(score_ranges, scores, color='skyblue') 
plt.xlabel('Score Range') 
plt.ylabel('Frequency') 
plt.title('Histogram of Students\' Scores') 
plt.show()
```

```python
# Conclusion: The histogram shows that most students scored between 31-40 marks, 
# indicating that this range is the most common score among the students.
```

```python
# Bivariate Analysis
# Scatter Plot A scatter plot is used to represent the relationship between two variables.
# Let's consider a dataset of students' scores in math and science tests.
# Bivariate Analysis - Scatter Plot 
math_scores = [45, 30, 50, 40, 35, 25] 
science_scores = [50, 35, 55, 45, 40, 30]
plt.figure(figsize=(10, 5)) 
plt.scatter(math_scores, science_scores, color='green')
plt.xlabel('Math Score') 
plt.ylabel('Science Score')
plt.title('Scatter Plot of Math vs Science Scores') 
plt.show()
```

```python
# Conclusion: The scatter plot shows a positive correlation between math and science scores, 
# indicating that students who score higher in math tend to score higher in science as well.
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Multivariate Analysis
# 3D Scatter Plot A 3D scatter plot is used to represent the relationship between three variables.
# Let's consider a dataset of students' scores in math, science, and English tests.
# Multivariate Analysis - 3D Scatter Plot 
english_scores = [55, 40, 60, 50, 45, 35] 
fig = plt.figure(figsize=(10, 5)) 
ax = fig.add_subplot(111, projection='3d')
ax.scatter(math_scores, science_scores, english_scores, color='red') 
ax.set_xlabel('Math Score') 
ax.set_ylabel('Science Score') 
ax.set_zlabel('English Score')
ax.set_title('3D Scatter Plot of Math, Science, and English Scores') 
plt.show()
```

```python
# Conclusion: The 3D scatter plot shows that there is a positive correlation among all three subjects. 
# Students who score higher in one subject 
# tend to score higher in the other two subjects as well.
```
